```python
# === Version-3 Code =================================
# Adds Level-1 upgrades:
#  - NVML power (avg_power_w) + TFLOPS/W
#  - BF16 support (skips if unsupported)
#  - Backend comparison: default | inductor | aot_eager
#  - Heatmap (TFLOPS) + Efficiency chart (TFLOPS/W)
# ==============================================================================

# 0) INSTALLS
import sys, subprocess, os, warnings
def pip_install(pkgs): subprocess.check_call([sys.executable, "-m", "pip", "install", "-q"] + pkgs)
pip_install(["torch", "pandas", "numpy", "matplotlib", "scikit-learn", "nvidia-ml-py3"])

# 1) IMPORTS
import time, csv, itertools, math
from datetime import datetime, timezone
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt

# NVML (power)
import pynvml

# 2) USER CONFIG
DEVICE_MODE   = "auto"                  # "auto" | "cuda" | "cpu"
SIZES         = [256, 512, 1024, 2048]  # M and N grid
K_VALUES      = [256, 512, 1024, 2048]  # K sweep
DTYPES        = ["fp32", "fp16", "bf16"]# bf16 will auto-skip on unsupported GPUs (e.g., T4)
BACKENDS      = ["default", "inductor", "aot_eager"]  # best-effort; skips on failure
ITERS         = 30
WARMUP        = 5
USE_COMPILE   = True
ALLOW_TF32    = True
OUTFILE       = "results_v3.csv"

# Visuals
HEATMAP_DTYPE    = "fp16"
HEATMAP_K        = 1024
HEATMAP_FILENAME = "heatmap_v3.png"
HEATMAP_TITLE    = "TFLOPS heatmap (matmul, K=1024, FP16)"
EFF_PLOT_FILE    = "efficiency_v3.png"  # TFLOPS/W by backend (top sizes)
EFF_TOP_N        = 8                     # show top-N efficient points

# 3) HELPERS
def dtype_from_str(s):
    s = s.lower()
    if s in ("fp32", "float32"): return torch.float32
    if s in ("fp16", "float16"): return torch.float16
    if s in ("bf16","bfloat16"): return torch.bfloat16
    raise ValueError(f"Unknown dtype: {s}")

def bf16_supported_on_cuda(device):
    try:
        major, minor = torch.cuda.get_device_capability(device)
        return (major, minor) >= (8, 0)  # Ampere+ typical
    except Exception:
        return False

def can_use_dtype_on_device(dtype, device):
    if device.type == "cpu":
        # Keep CPU simple/reliable: fp32 yes; bf16 often okay; fp16 often slow/unsupported.
        return dtype in (torch.float32, torch.bfloat16)
    if device.type == "cuda":
        if dtype is torch.bfloat16:
            return bf16_supported_on_cuda(device)
        return True
    return False

def make_tensors(M, N, K, dtype, device):
    A = torch.randn(M, K, dtype=dtype, device=device)
    B = torch.randn(K, N, dtype=dtype, device=device)
    return (A, B)

def build_op():
    # basic GEMM
    def f(A, B): return A @ B
    return f

def compile_with_backend(f, backend):
    if backend == "default" or not USE_COMPILE:
        return f
    try:
        return torch.compile(f, backend=backend, fullgraph=False)
    except Exception:
        return f  # fallback silently if backend not available

def time_op(f, tensors, iters, device):
    if device.type == "cuda":
        starter, ender = torch.cuda.Event(enable_timing=True), torch.cuda.Event(enable_timing=True)
        torch.cuda.synchronize()
        starter.record()
        for _ in range(iters):
            _ = f(*tensors)
        ender.record()
        torch.cuda.synchronize()
        ms = starter.elapsed_time(ender)
        return ms / 1000.0
    else:
        t0 = time.perf_counter()
        for _ in range(iters):
            _ = f(*tensors)
        return time.perf_counter() - t0

def flops_gemm(M, N, K):
    return 2.0 * M * N * K

def safe_nvml_init():
    try:
        pynvml.nvmlInit()
        return True
    except Exception:
        return False

def read_avg_power_w(samples=4, sleep_s=0.02):
    # Average a few samples to smooth out spikes
    try:
        handle = pynvml.nvmlDeviceGetHandleByIndex(0)
        vals = []
        for _ in range(samples):
            vals.append(pynvml.nvmlDeviceGetPowerUsage(handle) / 1000.0)  # mW->W
            time.sleep(sleep_s)
        return float(np.mean(vals))
    except Exception:
        return float("nan")

# 4) DEVICE SETUP
if DEVICE_MODE == "auto":
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
else:
    device = torch.device(DEVICE_MODE)

if device.type == "cuda":
    try:
        torch.backends.cuda.matmul.allow_tf32 = bool(ALLOW_TF32)
    except Exception:
        pass
device_name = torch.cuda.get_device_name(0) if device.type == "cuda" else "cpu"
print(f"Using device: {device} ({device_name})")

nvml_ok = False
if device.type == "cuda":
    nvml_ok = safe_nvml_init()
    print("NVML power tracking:", "ENABLED" if nvml_ok else "DISABLED")

# 5) BENCHMARK SWEEP
header = ["timestamp","device","device_name","op","dtype","backend","compiled","tf32",
          "M","N","K","iters","elapsed_s","tflops","avg_power_w","tflops_per_watt"]

first_write = not os.path.exists(OUTFILE)
with open(OUTFILE, "a", newline="") as fcsv:
    w = csv.writer(fcsv)
    if first_write:
        w.writerow(header)

    op = "matmul"
    for dtype_str, M, N, K, backend in itertools.product(DTYPES, SIZES, SIZES, K_VALUES, BACKENDS):
        # dtype
        try:
            dtype = dtype_from_str(dtype_str)
        except Exception:
            continue
        if not can_use_dtype_on_device(dtype, device):
            continue

        # tensors
        try:
            tensors = make_tensors(M, N, K, dtype, device)
        except RuntimeError:
            continue  # OOM etc.

        # function + backend
        f_base = build_op()
        f = compile_with_backend(f_base, backend)

        # Warmup
        for _ in range(WARMUP):
            _ = f(*tensors)
        if device.type == "cuda":
            torch.cuda.synchronize()

        # Power pre-sample
        pre_power = read_avg_power_w(samples=3) if (nvml_ok and device.type=="cuda") else float("nan")

        # Timed run
        elapsed = time_op(f, tensors, ITERS, device)

        # Power post-sample
        post_power = read_avg_power_w(samples=3) if (nvml_ok and device.type=="cuda") else float("nan")
        # Average (ignore NaN if NVML off)
        if math.isnan(pre_power) or math.isnan(post_power):
            avg_power_w = float("nan")
        else:
            avg_power_w = (pre_power + post_power) / 2.0

        tflops = flops_gemm(M, N, K) / (elapsed * 1e12)
        tflops_per_w = (tflops / avg_power_w) if (avg_power_w and not math.isnan(avg_power_w) and avg_power_w > 0) else float("nan")

        row = [datetime.now(timezone.utc).isoformat(), device.type, device_name, op, dtype_str,
               backend, int(USE_COMPILE and backend!="default"), int(ALLOW_TF32 if device.type=="cuda" else 0),
               M, N, K, ITERS, f"{elapsed:.6f}", f"{tflops:.6f}",
               (None if math.isnan(avg_power_w) else f"{avg_power_w:.3f}"),
               (None if math.isnan(tflops_per_w) else f"{tflops_per_w:.6f}")]
        w.writerow(row); fcsv.flush()

        print(f"[{op}] dtype={dtype_str:<4} backend={backend:<9} "
              f"M={M:4} N={N:4} K={K:4}  time={elapsed:.4f}s  "
              f"TFLOPS={tflops:.3f}  "
              f"PowerW={'n/a' if math.isnan(avg_power_w) else f'{avg_power_w:.1f}'}  "
              f"TF/W={'n/a' if math.isnan(tflops_per_w) else f'{tflops_per_w:.4f}'}")

print(f"\n Wrote results to {OUTFILE}")

# 6) VISUALS
df = pd.read_csv(OUTFILE)

# Heatmap: FP16, K fixed, backend=default (to compare apples-to-apples)
def render_heatmap(df, dtype_substr, K, out_file, title=None):
    dff = df[(df["op"]=="matmul") & (df["K"]==K)]
    dff = dff[dff["dtype"].str.contains(dtype_substr, case=False, na=False)]
    # Prefer default backend for heatmap
    if "backend" in dff.columns:
        dff = dff[dff["backend"].fillna("default")=="default"]
    if dff.empty:
        print("No data to plot for heatmap. Adjust HEATMAP_* config or rerun.")
        return
    pv = dff.pivot_table(index="M", columns="N", values="tflops", aggfunc="mean")
    plt.figure(figsize=(8,6))
    plt.imshow(pv.values, aspect="auto")
    plt.xticks(range(len(pv.columns)), pv.columns, rotation=45)
    plt.yticks(range(len(pv.index)), pv.index)
    plt.xlabel("N"); plt.ylabel("M")
    plt.title(title or f"TFLOPS heatmap (K={K}, dtype~{dtype_substr})")
    plt.colorbar(label="TFLOPS")
    plt.tight_layout(); plt.savefig(out_file, dpi=150)
    print(f" Saved heatmap to {out_file}")

render_heatmap(df, HEATMAP_DTYPE, HEATMAP_K, HEATMAP_FILENAME, HEATMAP_TITLE)

# Efficiency chart: top-N TFLOPS/W across backends (filter to large sizes for fairness)
def plot_efficiency(df, out_file, top_n=EFF_TOP_N):
    dff = df.copy()
    # keep rows with valid efficiency
    dff = dff[pd.to_numeric(dff["tflops_per_watt"], errors="coerce").notna()]
    if dff.empty:
        print("No TFLOPS/W data to plot (power readings unavailable).")
        return
    dff["tflops_per_watt"] = dff["tflops_per_watt"].astype(float)
    # Prefer larger sizes to avoid tiny-workload noise
    dff = dff[(dff["M"]>=512) & (dff["N"]>=512) & (dff["K"]>=512)]
    if dff.empty:
        print("No large-size efficiency data; showing all.")
        dff = df[pd.to_numeric(df["tflops_per_watt"], errors="coerce").notna()].copy()
        dff["tflops_per_watt"] = dff["tflops_per_watt"].astype(float)
    best = dff.sort_values("tflops_per_watt", ascending=False).head(top_n)
    labels = [f"{r['dtype']}/{r['backend']}\n{int(r['M'])}x{int(r['N'])}x{int(r['K'])}" for _, r in best.iterrows()]
    vals   = best["tflops_per_watt"].values
    plt.figure(figsize=(max(8, top_n*0.9), 6))
    plt.bar(range(len(vals)), vals)
    plt.xticks(range(len(vals)), labels, rotation=45, ha="right")
    plt.ylabel("TFLOPS per Watt")
    plt.title("Top Efficiency Runs (TFLOPS/W) by Dtype/Backend/Size")
    plt.tight_layout(); plt.savefig(out_file, dpi=150)
    print(f" Saved efficiency chart to {out_file}")

plot_efficiency(df, EFF_PLOT_FILE, top_n=EFF_TOP_N)

# Inline preview (if in notebook)
try:
    from IPython.display import Image, display
    if os.path.exists(HEATMAP_FILENAME): display(Image(filename=HEATMAP_FILENAME))
    if os.path.exists(EFF_PLOT_FILE):    display(Image(filename=EFF_PLOT_FILE))
except Exception:
    pass

# 7) QUICK PEEK AT CSV
print("\nFirst rows of results_v3.csv:")
from IPython.display import display
display(df.head(10))
```
