```python
# === Version-1 Code =====================================
# 1) Installs deps
# 2) Defines a PyTorch GEMM micro-benchmark (matmul/linear/bmm)
# 3) Runs a configurable sweep (M,N,K,dtype,ops,compile,TF32)
# 4) Saves results to results.csv
# 5) Renders a TFLOPS heatmap (matplotlib)
# 6) Prints the head of results.csv

# -----------------------------
# 0) INSTALL DEPENDENCIES
# -----------------------------
import sys, subprocess, os
def pip_install(pkgs):
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q"] + pkgs)

pip_install(["torch", "pandas", "numpy", "matplotlib"])

# -----------------------------
# 1) IMPORTS
# -----------------------------
import time, csv, itertools
from datetime import datetime
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt

# -----------------------------
# 2) USER CONFIG (edit here)
# -----------------------------
DEVICE_MODE   = "auto"           # "auto" | "cuda" | "cpu"
SIZES         = [128, 256, 512]  # M and N grid
K_VALUES      = [128, 256, 512]  # K sweep
DTYPES        = ["fp32", "fp16", "bf16"]  # will auto-filter unsupported
OPS           = ["matmul", "linear", "bmm"]
ITERS         = 20
WARMUP        = 5
BATCH_FOR_BMM = 4
USE_COMPILE   = True             # torch.compile best-effort
ALLOW_TF32    = True             # TF32 for CUDA matmul if supported
OUTFILE       = "results.csv"

# Heatmap rendering config (post-run)
HEATMAP_OP       = "matmul"
HEATMAP_DTYPE    = "fp16"        # "fp32"|"fp16"|"bf16" (must exist in results)
HEATMAP_K        = 256           # one value from K_VALUES
HEATMAP_FILENAME = "heatmap.png"
HEATMAP_TITLE    = "TFLOPS heatmap (matmul, K=256, fp16)"

# -----------------------------
# 3) HELPERS
# -----------------------------
def dtype_from_str(s):
    s = s.lower()
    if s in ("fp32","float32"): return torch.float32
    if s in ("fp16","float16"): return torch.float16
    if s in ("bf16","bfloat16"): return torch.bfloat16
    raise ValueError(f"Unknown dtype: {s}")

def can_use_dtype_on_device(dtype, device):
    if device.type == "cpu":
        # Colab CPU: fp32, bf16 generally safe; fp16 is hit/miss (MKLDNN)
        if dtype is torch.float16:
            try:
                return torch.backends.mkldnn.is_available()
            except Exception:
                return False
        return dtype in (torch.float32, torch.bfloat16)
    if device.type == "cuda":
        # fp16 widely supported; bf16 on newer GPUs; let it try, fallback if fails on alloc
        if dtype is torch.bfloat16:
            try:
                major, minor = torch.cuda.get_device_capability(device)
                return (major, minor) >= (8, 0)  # Ampere+ typically
            except Exception:
                return False
        return True
    return False

def make_tensors(op, M, N, K, dtype, device, batch):
    if op == "matmul":
        A = torch.randn(M, K, dtype=dtype, device=device)
        B = torch.randn(K, N, dtype=dtype, device=device)
        return (A, B)
    elif op == "linear":
        X = torch.randn(M, K, dtype=dtype, device=device)
        W = torch.randn(N, K, dtype=dtype, device=device)  # F.linear uses X @ W.T
        return (X, W)
    elif op == "bmm":
        batch = batch if batch and batch > 0 else 4
        A = torch.randn(batch, M, K, dtype=dtype, device=device)
        B = torch.randn(batch, K, N, dtype=dtype, device=device)
        return (A, B)
    else:
        raise ValueError(op)

def run_once(op, compiled=False):
    if op == "matmul":
        f = lambda A, B: A @ B
    elif op == "linear":
        import torch.nn.functional as F
        f = lambda X, W: F.linear(X, W)  # W is [N, K]
    elif op == "bmm":
        f = lambda A, B: torch.bmm(A, B)
    else:
        raise ValueError(op)
    if compiled:
        try:
            f = torch.compile(f, fullgraph=False)  # best-effort
        except Exception:
            pass
    return f

def time_op(f, tensors, iters, device):
    if device.type == "cuda":
        starter, ender = torch.cuda.Event(enable_timing=True), torch.cuda.Event(enable_timing=True)
        torch.cuda.synchronize()
        starter.record()
        for _ in range(iters):
            _ = f(*tensors)
        ender.record()
        torch.cuda.synchronize()
        ms = starter.elapsed_time(ender)
        return ms / 1000.0
    else:
        t0 = time.perf_counter()
        for _ in range(iters):
            _ = f(*tensors)
        return time.perf_counter() - t0

def flops_gemm(M, N, K):
    # 2*M*N*K FLOPs
    return 2.0 * M * N * K

# -----------------------------
# 4) DEVICE SETUP
# -----------------------------
if DEVICE_MODE == "auto":
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
else:
    device = torch.device(DEVICE_MODE)

if device.type == "cuda":
    try:
        torch.backends.cuda.matmul.allow_tf32 = bool(ALLOW_TF32)
    except Exception:
        pass

device_name = torch.cuda.get_device_name(0) if device.type == "cuda" else "cpu"
print(f"Using device: {device} ({device_name})")

# -----------------------------
# 5) BENCHMARK SWEEP
# -----------------------------
header = ["timestamp","device","device_name","op","dtype","compiled","tf32",
          "batch","M","N","K","iters","elapsed_s","tflops"]

first_write = not os.path.exists(OUTFILE)
with open(OUTFILE, "a", newline="") as fcsv:
    w = csv.writer(fcsv)
    if first_write:
        w.writerow(header)

    for op, dtype_str, M, N, K in itertools.product(OPS, DTYPES, SIZES, SIZES, K_VALUES):
        try:
            dtype = dtype_from_str(dtype_str)
        except Exception:
            continue
        if not can_use_dtype_on_device(dtype, device):
            continue

        # Create tensors
        try:
            tensors = make_tensors(op, M, N, K, dtype, device, BATCH_FOR_BMM)
        except RuntimeError:
            # OOM or unsupported allocation â€“ skip
            continue

        # Warmup
        f = run_once(op, compiled=USE_COMPILE)
        for _ in range(WARMUP):
            _ = f(*tensors)
        if device.type == "cuda":
            torch.cuda.synchronize()

        # Timed
        elapsed = time_op(f, tensors, ITERS, device)
        tflops = flops_gemm(M, N, K) / (elapsed * 1e12)

        row = [datetime.utcnow().isoformat(), device.type, device_name, op, dtype_str,
               int(USE_COMPILE), int(ALLOW_TF32 if device.type=="cuda" else 0),
               BATCH_FOR_BMM if op=="bmm" else "",
               M, N, K, ITERS, f"{elapsed:.6f}", f"{tflops:.6f}"]
        w.writerow(row)
        fcsv.flush()
        print(f"[{op:<6}] dtype={dtype_str:<5} M={M:4} N={N:4} K={K:4}  "
              f"time={elapsed:.4f}s  TFLOPS={tflops:.3f}")

print(f"\nâœ… Wrote results to {OUTFILE}")

# -----------------------------
# 6) VISUALIZE HEATMAP
# -----------------------------
df = pd.read_csv(OUTFILE)

def render_heatmap(df, op, dtype_substr, K, out_file, title=None):
    dff = df[(df["op"]==op) & (df["K"]==K)]
    # dtype column contains strings like 'fp16', 'fp32', etc. Keep substring match
    mask = dff["dtype"].str.contains(dtype_substr, case=False, na=False)
    dff = dff[mask]
    if dff.empty:
        print("No data to plot. Adjust HEATMAP_* config.")
        return
    pv = dff.pivot_table(index="M", columns="N", values="tflops", aggfunc=np.mean)
    plt.figure(figsize=(8,6))
    plt.imshow(pv.values, aspect="auto")
    plt.xticks(range(len(pv.columns)), pv.columns, rotation=45)
    plt.yticks(range(len(pv.index)), pv.index)
    plt.xlabel("N")
    plt.ylabel("M")
    plt.title(title or f"TFLOPS heatmap (op={op}, K={K}, dtype~{dtype_substr})")
    plt.colorbar(label="TFLOPS")
    plt.tight_layout()
    plt.savefig(out_file, dpi=150)
    print(f"ðŸ“ˆ Saved heatmap to {out_file}")

render_heatmap(df, HEATMAP_OP, HEATMAP_DTYPE, HEATMAP_K, HEATMAP_FILENAME, HEATMAP_TITLE)

# Show in cell
if os.path.exists(HEATMAP_FILENAME):
    from IPython.display import Image, display
    display(Image(filename=HEATMAP_FILENAME))

# -----------------------------
# 7) Displays the first few rows of results.csv
# -----------------------------
import pandas as pd
df = pd.read_csv("results.csv")
from IPython.display import display
print("\nFirst rows of results.csv:")
display(df.head())
# =============================================================================
```
